$$PARTE \quad 1D $$

```python
import numpy as np
import matplotlib.pyplot as plt

r_lista = [1, 2, 4]
t_punti = np.linspace(0, 10, 5000)

plt.figure(figsize=(12,6))

for r in r_lista:
    
    t_validi = []  ##per non avere le linee a 0 
    u_validi = []
    
    for t in t_punti:
        tau = t - r
        if tau >= 0:
            valore_u = np.exp(-tau) * np.sin(4 * tau) / np.sqrt(r)
            t_validi.append(t)
            u_validi.append(valore_u)
    
    #plot dei _validi
    plt.plot(t_validi, u_validi, label=f"r = {r}", lw=2)

plt.title("Dipendenza u(t)")
plt.xlabel("Tempo [t]")
plt.ylabel("u(t)")
plt.axhline(0, color='black', lw=1, ls="--")
plt.xlim(0, 10) 
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
    
```

$$ INTERFERENZA $$


```python
import numpy as np
import matplotlib.pyplot as plt

L = 12
N = 250

x = np.linspace(-L/2, L/2, N)
y = np.linspace(-L/2, L/2, N)

X,Y = np.meshgrid(x, y)

R1 = np.sqrt(X**2 + (Y-2)**2)  ##distanze per le due onde
R2 = np.sqrt(X**2 + (Y+2)**2)

def calcolo_u (t, R, v=1):  ##stessa funzione di prima la uso per onda2
    tau = t - R/v
    mask = (tau >=0 )
    u = np.where(mask, np.exp(-tau) * np.sin(4 * tau) / np.sqrt(R + 1e-10), 0.0) #aggiungo un piccolo valore a R per non far dividere per 0
    return u

def onda_due (t, R1, R2):
    u1 = calcolo_u(t, R1)
    u2 = calcolo_u(t, R2)
    return u1 + u2

##generazione immagini praticamente uguale a prima#
nframe = 200
dt = 0.05
for i in range(nframe):
    t_adesso = i * dt
    
    # Usiamo la nuova funzione con le due sorgenti
    u_totale = onda_due(t_adesso, R1, R2)
    
    plt.figure(figsize=(6, 6))
    plt.imshow(u_totale, 
               extent=[-L/2, L/2, -L/2, L/2], origin='lower', cmap='RdBu', vmin=-2, vmax=2)  
    ##ho cambiato i vmax e min perchè i picchi potrebbero superare 1?#
    plt.title(f"Interferenza a t = {t_adesso:.2f}")
    plt.savefig(f"interf{i:03d}.png", dpi=80, bbox_inches='tight')
    plt.close()
print("finito tutto 2")

```

$$MOTO \quad BROWNIANO --PARTE \space2$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
N = 1000
T = 2000
rng = np.random.default_rng()


dx = rng.normal(loc=0, scale=1, size=(N, T))
dy = rng.normal(loc=0, scale=1, size=(N, T))

x = np.cumsum(dx, axis=1) #somma di tutti gli spostamenti lungo x e y
y = np.cumsum(dy, axis=1) #quindi le righe della matrice(axis 1) 

r2 = x**2+y**2
media_r2 = np.mean(r2, axis=0) #media di r che è funz del tempo quindi (axis 0) lungo colonne 
t = np.linspace(1, T, T)#asse tempi con 2000 step 

#grafico
plt.figure(figsize=(10,10))
plt.loglog(t,media_r2,label="dati simulati")
#verifica retta con m=1
#rettath = t * (media_r2[0]/t[0]) #retta con m=1 per la verifica, moltiplico per quel termine così inizia allo stesso punto
#plt.loglog(t,rettath,alpha=0.7,ls="--",label="retta teorica manuale")
coeff = np.polyfit(np.log(t), np.log(media_r2), 1)##coeff del grafico

rettath2 = t*4*(np.exp(coeff[1])/4)
plt.loglog(t, rettath2,alpha=0.7,ls="--",label="retta teorica col D")

print(f"i coefficienti della retta sono (m, q) = ({coeff[0]:.2f}, {coeff[1]:.2f}) e il D è uguale a {np.exp(coeff[1])/4:.2f}")
plt.xlabel("T [step 1-2000]")
plt.ylabel("⟨r²(t)⟩ [u.a²]")
plt.title("⟨r²(t)⟩  v.s  T")
plt.legend()

#### grafico traiettorie ###
plt.figure(figsize=(8,8))
for i in range (3):
    plt.plot(x[i , :], y[i , :], label=f"particella{i+1}",ls="-")
    plt.plot(x[i,-1], y[i,-1], 'rs', markersize = 4)

plt.plot(0, 0, 'go', markersize=10, label='Partenza (Origine)')#in realtà la posizione iniziale per come l'ho calcolata non è 0 0 per tutte le particelle ma i dx e dy sono piccoli e graficamente non si nota
                                                               #per correttezza bisognerebbe aggiungere una colonna di 0 alle matrici x e y lungo l'asse 1
plt.axis('equal') 
plt.grid(True, alpha=0.3)
plt.title('Traiettorie individuali di 5 particelle')
plt.xlabel('Posizione x (u.a.)')
plt.ylabel('Posizione y (u.a.)')
plt.legend()

plt.show()

print("finito")


```
